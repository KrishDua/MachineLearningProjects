```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
irisdata = datasets.load_iris();
x = irisdata.data;
y = irisdata.target;
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier();
clf.fit(x,y);
```

```python
from sklearn.tree import export_graphviz
import pydotplus
dotdata = export_graphviz(clf,class_names = irisdata.target_names,feature_names = irisdata.feature_names,
                          filled = True,
                          rounded = True,
                          special_characters = True);
graph = pydotplus.graph_from_dot_data(dotdata);
graph.write_pdf("IrisDataDecisionTree.pdf");
```

```python
def entropy(y):
    # calculating entropy
    uniqueclasses,counts = np.unique(y,return_counts = True);
    probs = counts/counts.sum();
    return -np.sum(probs*np.log2(probs + 1e-9));
def infogain(x,y,featureidx):
    totalentropy = entropy(y);
    uniqueclasses,counts = np.unique(x[:,featureidx],return_counts = True);
    infosplit = 0;
    for i in range(len(uniqueclasses)):
        prob = counts[i]/counts.sum();
        subsetentropy = entropy(y[x[:,featureidx] == uniqueclasses[i]]);
        infosplit += (prob)*subsetentropy;
    return totalentropy - infosplit;
def gainratio(x,y,featureidx):
    ig = infogain(x,y,featureidx);
    splitinfo = entropy(x[:,featureidx]);
    return ig/(splitinfo + 1e-9);
def buildtree(x,y,featurenames,level = 0):
    uniqueclasses,counts = np.unique(y,return_counts = True);
    print("Level ",level);
    for cls,count in zip(uniqueclasses,counts):
        print("Count of ",cls," = ",count);
    print("Current Entropy is  = {:.12f}".format(entropy(y)));
    if(len(uniqueclasses) == 1):
        print("Reached Leaf Node");
        return;
    if(x.shape[1] == 0):
        print("no features left to split");
        return;
    gainratios = [gainratio(x,y,i) for i in range(x.shape[1])];
    bestfeature = np.argmax(gainratios);
    print("Splitting on ",featurenames[bestfeature]," with gain ratio {:.12f}".format(gainratios[bestfeature]));
    values = np.unique(x[:,bestfeature]);
    for value in values:
        x1 = x[x[:,bestfeature] == value];
        y1 = y[x[:,bestfeature] == value];
        buildtree(x1,y1,featurenames,level = level + 1);
```

```python
featurenames = irisdata.feature_names;
buildtree(x,y,featurenames);
```
